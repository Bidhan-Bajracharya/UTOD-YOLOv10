```python
from ultralytics import YOLOv10
import matplotlib.pyplot as plt
import cv2

model = YOLOv10.from_pretrained('jameslahm/yolov10s')

# Pick your own sources or keep the built-in sample images
sources = [
    r"D:\UTOD\UTOD-YOLOv10\ultralytics\assets\bus.jpg",
    r"D:\UTOD\UTOD-YOLOv10\ultralytics\assets\zidane.jpg"
]

results = model.predict(
    source=sources,         # file, folder, glob, or 0 for webcam
    conf=0.25,
    device=0,
    save=True,              # saves annotated images
    project=r"D:\UTOD\yolov10\runs",  # control where files go
    name="demo",            # subfolder name
    exist_ok=True
)

# Display annotated images inline
for r in results:
    im_bgr = r.plot()            # annotated image as BGR np.array
    im_rgb = cv2.cvtColor(im_bgr, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(8,6))
    plt.imshow(im_rgb)
    plt.axis('off')
    plt.title(r.path)
    plt.show()

print("Saved to:", results[0].save_dir)  # e.g., D:\UTOD\yolov10\runs\demo
```
